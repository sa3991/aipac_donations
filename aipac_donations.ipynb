```python
import pandas as pd
df = pd.read_csv("aipac.csv")
%matplotlib inline
#This data is from the Federal Election Commission's campaign finance data receipts from 2023-2024. Link: https://www.fec.gov/data/receipts/?data_type=processed&contributor_name=C00797670&two_year_transaction_period=2024
```

```python
#I'm removing extra columns from the data so it's easier to review
df_clean = df[['committee_name', 'report_year', 'report_type', 'entity_type', 'contributor_name','contribution_receipt_amount','contributor_aggregate_ytd','receipt_type_full','fec_election_type_desc','fec_election_year','contribution_receipt_date','pdf_url','transaction_id']]
```

```python
#I'm removing donations from individuals to only show donations from PACs
df_clean_final = df_clean[df_clean.entity_type != 'IND']
```

```python
df_clean_final
```

```python
type(df_clean_final['contribution_receipt_date'][1])
```

```python
df_date = df_clean_final
```

```python
#I'm filtering the data to only see donations from this year
df_date_final = df_date[df_date['contribution_receipt_date'] > '2023-12-31']
```

```python
#I'm looking at the oldest donation in this dataset
df_date_final['contribution_receipt_date'].min()
```

```python
#I'm looking at the most recent donation in this dataset
df_date_final['contribution_receipt_date'].max()
```

```python
#I'm dropping any duplicate donations that have the same transaction ID
df_clean_final_2 = df_clean_final.drop_duplicates(subset=['transaction_id'])
```

```python
#I'm creating a data frame that comnbines the sum of donations for each candidate
df_sum = df_date_final.groupby('committee_name')['contribution_receipt_amount'].sum().reset_index()
```

```python
df_sum
```

```python
#I organized the dataset to see which candidates AIPAC gave the most money to
df_table = df_sum.sort_values(by='contribution_receipt_amount',ascending=False).head(5)
```
