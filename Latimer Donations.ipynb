```python
import pandas as pd
df = pd.read_csv("latimer_donations.csv", low_memory=False)
#This data is from the Federal Election Commission's campaign finance data receipts from 2023-2024. Link: https://www.fec.gov/data/receipts/?data_type=processed&committee_id=C00859041&two_year_transaction_period=2024
```

```python
df.head()
```

```python
df.tail()
```

```python
#I'm removing extra columns from the data so it's easier to review
df_clean = df[['committee_name', 'report_year', 'report_type', 'entity_type', 'contributor_name','contribution_receipt_amount','contributor_aggregate_ytd','receipt_type_full','fec_election_type_desc','fec_election_year','contribution_receipt_date','pdf_url','transaction_id']]
```

```python
df_clean
```

```python
#I'm filtering the data to only see donations from AIPAC
df_aipac = df_clean[df_clean['contributor_name'].str.contains('AMERICAN ISRAEL PUBLIC AFFAIRS COMMITTEE')]
```

```python
df_aipac
```

```python
#I'm filtering the data to only see donations from this year
df_date = df_clean[df_clean['contribution_receipt_date'] > '2023-12-31']
```

```python
#I'm looking at the oldest donation in this dataset
df_date['contribution_receipt_date'].min()
```

```python
#I'm looking at the most recent donation in this dataset
df_date['contribution_receipt_date'].max()
```

```python
#I'm creating a dataset that comnbines the sum of donations for each candidate
df_table = df_date.groupby('contributor_name')['contribution_receipt_amount'].sum().reset_index()
```

```python
df_table
```

```python
df_final_table = df_table.sort_values(by='contribution_receipt_amount', ascending=False).head(4)
```

```python
#I organized the dataset to see what PACs gave the most money to Latimer
df_final_table
```

```python
#I'm exporting the table as a csv file so I can create a graph
df_final_table.to_csv('latimer_donations_table_final', index=False)
```
